```python
import matplotlib.pyplot as plt
import numpy as np
import os
#change directory to save output here
directory = r"D:\Users\Nick\Desktop\Everything\Uni\Honours Project\Plots\MiscExamples/"
plt.rcParams["savefig.directory"] = os.chdir(os.path.dirname(directory))

X=np.array([-4,-3,3,4])
Y=np.array([-2,-1,0,1,2])
X2=np.array([16,9,9,16])
Y2=np.array([4,1,0,1,4])
```

```python
plt.rcParams.update({'font.size': 15})
plt.figure(figsize = (5,5))
for i in X:
    plt.plot(i,0,"go")
for ii in Y:
    plt.plot(ii,0,"bo")

plt.tight_layout()
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.grid()
plt.ylim(-2,17)
plt.show
plt.savefig("lindata.pdf",format="pdf",bbox_inches="tight")
```

```python
z1 = np.column_stack((X,X2))
z2 = np.column_stack((Y,Y2))
```

```python
n=0
plt.figure(figsize = (5,5))
for i in z1:
    plt.plot(z1[n,0],z1[n,1],"go")
    n+=1
n=0
for i in z2:
    plt.plot(z2[n,0],z2[n,1],"bo")
    n+=1
plt.plot([-4,4],[6,6],"ro-",lw=2)
plt.ylabel("$X_1^2$")
plt.xlabel("$X_1$")
plt.tight_layout()
plt.grid()
plt.ylim(-2,17)
plt.show
plt.savefig("sepdata.pdf",format="pdf",bbox_inches="tight")
```
